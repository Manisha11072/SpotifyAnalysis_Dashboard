```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import time

# Load your Spotify data
file_path = '/content/Spotify Most Streamed Songs.csv'
spotify_df = pd.read_csv(file_path)

# Set up Spotify API credentials
client_id = '86e61c62665e420ab74d8e77c2c7352a'
client_secret = '5f3305390fd8439d818fd4bc788b2daa'

# Authenticate with Spotify
auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager)

# Function to get Spotify URLs for track and album
def get_spotify_urls(track_name, artist_name):
    try:
        # Search for the track on Spotify
        result = sp.search(q=f"track:{track_name} artist:{artist_name}", type='track', limit=1)
        if result['tracks']['items']:
            track = result['tracks']['items'][0]
            track_url = track['external_urls']['spotify']
            album_url = track['album']['external_urls']['spotify']
            cover_url = track['album']['images'][0]['url'] if track['album']['images'] else None
            return track_url, album_url, cover_url
        else:
            return None, None, None
    except Exception as e:
        print(f"Error retrieving data for {track_name} by {artist_name}: {e}")
        return None, None, None

# Add new columns for URLs
spotify_df['track_url'] = None
spotify_df['album_url'] = None
spotify_df['cover_url'] = None

# Populate URLs for each song
for index, row in spotify_df.iterrows():
    track_url, album_url, cover_url = get_spotify_urls(row['track_name'], row['artist(s)_name'])
    spotify_df.at[index, 'track_url'] = track_url
    spotify_df.at[index, 'album_url'] = album_url
    spotify_df.at[index, 'cover_url'] = cover_url
    time.sleep(0.1)  # To avoid rate limits

# Save the updated dataset
spotify_df.to_csv('Updated_Spotify_Most_Streamed_Songs_with_URLs.csv', index=False)
print("Updated dataset saved as 'Updated_Spotify_Most_Streamed_Songs_with_URLs.csv'")
```
